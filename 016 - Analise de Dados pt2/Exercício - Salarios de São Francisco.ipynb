**Quantidade de funcionários e por JobTitle**<br>
**Total gasto e por agência**<br>
**Média de salário por JobTitle**<br>
**Total de horas extras**<br>

```python
Func = sal['EmployeeName'].unique()
qtdefunc = len(Func)
print(f'Quantidade de funcionários únicos: {qtdefunc}')

print ("_____________________________________________________________")

jobFunc = sal.groupby("JobTitle")["Id"].count()
print(jobFunc)

print ("_____________________________________________________________")

totalSal = sal['OvertimePay'].sum()
print(f'A maior quantidade de horas extras foi: ${totalSal}')
```

##### Qual é o "BasePay"?

```python
midiasal = sal['BasePay'].mean()

mes = midiasal / 12
print(f'O salário base médio é: ${mes:.2f}')
```

##### Qual é a maior quantidade de "OvertimePay" no conjunto de dados

```python
MaiorOvertime = sal['OvertimePay'].max()
print(f'A maior quantidade de "OvertimePay" no conjunto de dados é: ${MaiorOvertime}')
```

**Qual é o cargo de JOSEPH DRISCOLL? Nota: use todas as maiúsculas, caso contrário você pode obter uma resposta que não coincida (há também um Joseph Driscoll com minúsculas).**

```python
CargoJoseph = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle'].values[0]
print(f'O cargo de JOSEPH DRISCOLL é: {CargoJoseph}')
```

**Quanto JOSEPH DRISCOLL ganha (incluindo benefícios)?**

```python
PaymentsJoseph = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits'].values[0]

print(f'O total de pagamento de JOSEPH DRISCOLL é: ${PaymentsJoseph}')
```

**Qual o nome da pessoa mais bem paga (incluindo benefícios)?**

```python
maxSal = sal['TotalPayBenefits'].max()
empMaxSal = sal[sal['TotalPayBenefits'] == maxSal]['EmployeeName'].values[0]
print(f'O funcionário com o maior salário total é: {empMaxSal} com um salário de: ${maxSal}')  
```

**Qual o nome da pessoa paga mais baixa (incluindo benefícios)? Você percebe algo estranho sobre o quanto ele ou ela é paga?**

```python
minSal = sal['TotalPayBenefits'].min()
empMinSal = sal[sal['TotalPayBenefits'] == minSal]['EmployeeName'].values[0]
print(f'O funcionário com o maior salário total é: {empMinSal} com um salário de: ${minSal}')  
```

**Qual foi a média (média) BasePay de todos os funcionários por ano? (2011-2014)?**

```python
MediaAnoSalBase = sal.groupby('Year')['BasePay'].mean()

print("Média anual do salário base:")
print(MediaAnoSalBase)
```

**Quantos títulos de trabalho únicos existem?**

```python
TrabalhosUnicos = sal['JobTitle'].nunique()
print(f'Número de trabalhos únicos: {TrabalhosUnicos}')
```

**Quais são os 5 principais empregos mais comuns?**

```python
empregosTop5 = sal["JobTitle"].value_counts().head(5)
print("Top 5 empregos mais comuns:")
print(empregosTop5)
```

**Quantos Títulos de Trabalho foram representados por apenas uma pessoa em 2013? (Por exemplo, títulos de trabalho com apenas uma ocorrência em 2013?)**

```python
x = sal[sal['Year'] == 2013]
onejob= sum(x['JobTitle'].value_counts() == 1)
print(f"teve no total {onejob} trabalhos que foram exercidos por apenas uma pessoa em 2013")
```
