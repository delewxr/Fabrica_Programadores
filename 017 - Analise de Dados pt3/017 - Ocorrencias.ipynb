## Ocorrencias

```python
import pandas as pd
```

```python
df = pd.read_csv('OCORRENCIAS.csv', sep=';', encoding='latin1')
display(df.head())
```

```python
for colunas in df.columns:
    print(colunas)
```

## Quantas Ocorrencias teve no ano de 2025

```python
## Quantas Ocorrencias teve no ano de 2025
print('No ano de 2025 houve', df.groupby('ANO_OCORRENCIA').size()[2025], 'Ocorrencias')
```

# KPI's
## 1 Listar os municipios com mais ocorrencias
## listar os municipios com menos ocorrencias

```python
municipios = df.groupby('MUNICIPIO')["TOTAL"].sum()
print('Municipio com menos ocorrencias:', municipios)

print('____________________________________________________')

top5mais = municipios.nlargest(10)
display('Top 10 municipios com mais ocorrencias:', top5mais)

print('____________________________________________________')

top5menos = municipios.nsmallest(10)
display('Top 10 minucipios com menos ocorrencias:', top5menos)
```

# 2 Tipo de arma mais utilizado

```python
arma = df.groupby('ESPECIE_ARMA')["TOTAL"].sum()
armamais = arma.idxmax()
armamenos = arma.idxmin()
print(F'Tipo de arma mais utilizada: {armamais}')
print(f'Tipo de arma menos utilizada: {armamenos}')
```

# 3 Total de ocorrencias

```python
total_ocorrencias = df['TOTAL'].sum()
print(f'Total de ocorrencias: {total_ocorrencias}')
```

# 4 Histórico de ocoorencias durante o ano

```python
## quantas ocorrencias em media as cidades com mais e menos ocorrencias tiveram
mes = df.groupby('MES_OCORRENCIA')["TOTAL"].sum()
print(f"Segue o histórico de ocorrencias durante o ano")
print(mes)
```

# 5 Média de ocorrencias entre as cidades com mais e menos Habitantes

```python
## Quantas ocorrencias teve em média entre as cidades com mais e menos ocorrencias
cidadesMil = df.groupby('MAIS_1000_MIL_HAB')["TOTAL"].sum()

print(f'Total de ocorrencias: {total_ocorrencias}\n')

cidademilsim = cidadesMil.get("S")
cidademilnao = cidadesMil.get("N")

Conta1 = cidademilsim / total_ocorrencias
conta2 = cidademilnao / total_ocorrencias

print(f'Porcetagem de ocorrencias entre as cidades com mais de 1000 habitantes: {Conta1:.2%}\n')
print(f'Porcetagem de ocorrencias entre as cidades com menos de 1000 habitantes: {conta2:.2%}')
```
