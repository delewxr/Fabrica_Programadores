### Respondendo as perguntas

```python
import pandas as pd

#quanto foi vendido de cada cart√£o
df = pd.read_csv('Ecommerce_Purchases.csv')

vendas_por_cartao = df.groupby('CC Provider')['Purchase Price'].sum()
print(vendas_por_cartao)
```

```python
#quantidade de compras por cliente
clientes = df['Company'].unique()
resposta = len(clientes)

print(f'Vendemos para {resposta} clientes diferentes.')
```

```python
#em quantos idiomas diferentes foi comprado
idiomas = df['Language'].unique()
print(f'As compras foram feitas em {len(idiomas)} idiomas diferentes.')

for idioma in idiomas:
    print(idioma)
```

```python
#qual tipo de trabalho mais comprou
trabalho = df['Job'].max()

print(f'O tipo de trabalho que mais comprou foi: {trabalho}.')
```
