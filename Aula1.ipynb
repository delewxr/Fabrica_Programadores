```python
import pandas as pd

df = pd.read_csv('estoque.csv')
#ver primeiras linhas do dataframe
display(df.head(2))

print('\n________________________________\n')

#ver colunas do dataframe
#print(df.columns)

for i in df.columns:
    print(i)
```

```python
import pandas as pd

#ver uma coluna específica
print(df['marca'])

print('\n________________________________\n')

#ver várias colunas específicas
print(df[['marca','valor']])
```

```python
#agrupar os dados pela coluna marca
#quantos tipos de marcas diferentes existem no estoque
grupo_marca = df.groupby('modelo')["modelo"].count()
print(grupo_marca)

print('\n________________________________\n')

cores = df.groupby('cor')["cor"].count()
print(cores)
```

```python
#criar uma coluna nova

df['faturamento'] = df['quantidade'] * df['valor']

display(df.head())
```

```python
#qual o valor total desse estoque

total = df['faturamento'].sum()

print(f'O valor total do estoque é de R$ {total} reais.')

print('\n________________________________\n')

estoque = df['quantidade'].sum()
print(f'A quantidade total de itens no estoque é de {estoque} unidades.')

estoque_tipo = df.groupby('tipo')['quantidade'].sum()
print(estoque_tipo)

# valor total por marca

valor_marca = df.groupby('marca')['faturamento'].sum()
print(valor_marca)
```

```python
#filtros
#filtrar o estoque com veiculos com até 10 anos de uso
novos = df[df['ano'] >= 2015]
display(novos)

print('\n____________________________________________________________________________________\n')

#filtro - carregar apenas os carros

carros = df[df['tipo'] == 'Moto']
display(carros)
```

```python
# criar consulta de veiculos por cores

print('pesquise veiculos pela cor desejada')
cor = input('Digite a cor desejada: ')
consulta = df[df['cor'] == cor]
display(consulta)
```
